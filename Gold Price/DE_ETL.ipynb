```python
import pandas as pd

# 1. Start with 'r' for a raw string
# 2. Use the full absolute path, including the filename
gold_dataset = pd.read_csv(r'C:\Users\baska\Desktop\New Data_journey\Data\Youtube_Stackoverflow\Gold Price\gold_price_data_new1_2024.csv')

print("Data loaded successfully!")
print(gold_dataset.head())
```

```python
import sqlite3
```

```python
gold_dataset['Gold rate']=gold_dataset['Gold rate'].replace('₹','',regex=True).replace(',','',regex=True).astype(float)
```

```python
gold_dataset.head()
```

```python
conn = sqlite3.connect("gold_data.db")
```

```python
gold_dataset.to_sql("gold_prices", conn, if_exists="replace", index=False)
```

```python
query_result = pd.read_sql("SELECT * FROM gold_prices LIMIT 5;", conn)
print(query_result.head())
```
